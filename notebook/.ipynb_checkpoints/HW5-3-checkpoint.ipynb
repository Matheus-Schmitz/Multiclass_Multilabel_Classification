# Imports

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import warnings
warnings.filterwarnings("ignore")
```

## ISLR 10.7.2

```python
dissimilarity_matrix = np.matrix([[0.00, 0.30, 0.40, 0.70], 
                                  [0.30, 0.00, 0.50, 0.80],
                                  [0.40, 0.50, 0.00, 0.45],
                                  [0.70, 0.80, 0.45, 0.00]])
dissimilarity_matrix
```

### (a) Complete Linkage Dendogram

Code based on: https://stackabuse.com/hierarchical-clustering-with-python-and-scikit-learn/

```python
linked = linkage(dissimilarity_matrix, 'complete')

plt.figure(figsize=(10, 7))
dendrogram(linked, distance_sort='descending', show_leaf_counts=True)
plt.show()
```

```python
Z = linkage(dissimilarity_matrix, 'ward')
```
